```python
## Teams list
from nba_api.stats.static import teams 
import pandas as pd
team_df = pd.DataFrame(teams.get_teams())

print('Teams')
print(f'Shape: {team_df.shape}')
display(team_df.head())
```

```python
## Players list
from nba_api.stats.static import players 
import pandas as pd
player_df = pd.DataFrame(players.get_players())

print('All Players')
print(f'Shape: {player_df.shape}')
display(player_df.head())

## get active players by filtering on is_active
active_players = player_df[player_df['is_active'] == True]
print('Active Players')
print(f'Shape: {active_players.shape}')
display(active_players.head())
```

```python
## get all team data for last 20 years
from nba_api.stats.endpoints import leaguegamelog
import pandas as pd
# import time

def get_team_game_data(year, season_type_all_star):
    lg = leaguegamelog.LeagueGameLog(season=str(year), season_type_all_star=season_type_all_star)
    data_frame = lg.get_data_frames()[0]
    data_frame['SeasonTypeAllStar'] = season_type_all_star
    return data_frame

team_game_data = pd.DataFrame()
season_types = ['Regular Season', 'Pre Season', 'Playoffs', 'All-Star']

for year in range(2020, 2024):
    print(f'Getting data for {year}')
    for season_type in season_types:
        team_game_data = pd.concat([team_game_data, get_team_game_data(year, season_type)], ignore_index=True)

print('\nTeam Game Data')
print(f'Shape: {team_game_data.shape}')
display(team_game_data.head())

print('\nNOTE: Both season and season type are in the SEASON_ID column, so we can use that to filter')
display(team_game_data.groupby(['SEASON_ID', 'SeasonTypeAllStar']).size().head(5))
```
