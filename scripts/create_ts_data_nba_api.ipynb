```python
def get_min(time_str):
    """Get minutes from time (MM:SS)"""
    try:
        m, s = time_str.split(':')
        mins = float(m) + float(s)/60
    except:
        mins = float(time_str)
    return mins

df_game_detail2['MIN'] = df_game_detail2['MIN'].apply(get_min)
print(df_game_detail2['MIN'].dtype)
```

## Create timeseries data

```python
## Lag and average window functions
def get_all_lags(df, cols:t.List, group_cols:t.List, n_lags:int) -> t.Tuple[pd.DataFrame, t.List[str]]:
    df_ = df.copy()
    created_cols = []

    if n_lags > 0:
        start, stop, step = 1, n_lags + 1, 1
        prefix = 'lag_'
    else:
        start, stop, step = 0, n_lags, -1
        prefix = 'next_'
    
    # for each column in cols, get n_lags number of lags
    for col in cols:
        # for each lag, create a new column with the lagged value
        for i in range(start, stop, step):
            created_col = f'{prefix}{abs(i)}_{col}'
            created_cols.append(created_col)
            df_[created_col] = df_.groupby(group_cols)[col].shift(i)

    return df_, created_cols

def get_all_average_windows(df, cols:t.List, group_cols:t.List, 
            prefix:str, windows:np.array, shifts:np.array, min_periods:np.array
            ) -> t.Tuple[pd.DataFrame, t.List[str]]:
    df_ = df.copy()
    created_cols = []

    # for each column in cols, get average over a window of games
    for col in cols:
        # for each window, create a new column with the average value
        for w, s, mp in zip(windows, shifts, min_periods):
            created_col = f'{prefix}{s}_{s+w-1}_{col}'
            created_cols.append(created_col)
            df_[created_col] = df_.groupby(group_cols)[col].shift(s).rolling(w, min_periods=mp).mean()

    return df_, created_cols
```

```python
df_scoreboard_slim = df_scoreboard[['GAME_DATE_EST', 'GAME_ID', 'GAME_STATUS_TEXT', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'SEASON']]

## only where trim(GAME_STATUS_TEXT) == 'Final'
df_scoreboard_slim = df_scoreboard_slim[df_scoreboard_slim['GAME_STATUS_TEXT'].str.strip() == 'Final']

df_game_log_slim = df_game_log[['GAME_ID', 'TEAM_ID', 'PTS', 'FG_PCT', 'FG3_PCT', 'AST', 'REB', 'TOV', 'SEASON_ID']]

# First, merge the home team data
df_games = df_scoreboard_slim.merge(df_game_log_slim.add_suffix('_home'), how='left', left_on=['GAME_ID', 'HOME_TEAM_ID'], right_on=['GAME_ID_home', 'TEAM_ID_home'])

# Then, merge the visitor team data with different suffix
df_games = df_games.merge(df_game_log_slim.add_suffix('_away'), how='left', left_on=['GAME_ID', 'VISITOR_TEAM_ID'], right_on=['GAME_ID_away', 'TEAM_ID_away'])

## remove rows where any of [GAME_ID_home, TEAM_ID_home, GAME_ID_away, TEAM_ID_away] contain NaN
df_games = df_games.dropna(subset=['GAME_ID_home', 'TEAM_ID_home', 'GAME_ID_away', 'TEAM_ID_away'])

# ##seasons starting with 2 are regular season (e.g. 22000 is regular season 2020-2021)
start_with_2_mask = np.char.startswith(np.char.mod('%d', df_games['SEASON_ID_home']), '2')
df_games['regular_season'] = np.where(start_with_2_mask, 1, 0)

df_games.drop(['SEASON_ID_home', 'SEASON_ID_away', 'GAME_ID_home', 'GAME_ID_away'], axis=1, inplace=True)

## convert GAME_DATE_EST to datetime
df_games['GAME_DATE_EST'] = pd.to_datetime(df_games['GAME_DATE_EST'])

display(df_games)
```

```python
## 0. Transform games table from HOME_TEAM, AWAY_TEAM to TEAM, HOME/AWAY 
game_general_cols = ['GAME_DATE_EST', 'GAME_ID', 'SEASON', 'regular_season']
game_home_cols = ['TEAM_ID_home', 'PTS_home', 'FG_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home', 'TOV_home']
game_away_cols = ['TEAM_ID_away', 'PTS_away', 'FG_PCT_away', 'FG3_PCT_away', 'AST_away', 'REB_away', 'TOV_away']

team_stat_cols = ['TEAM_PTS', 'TEAM_FG_PCT', 'TEAM_FG3_PCT', 'TEAM_AST', 'TEAM_REB', 'TEAM_TOV']
opp_team_stat_cols = ['OPP_TEAM_PTS', 'OPP_TEAM_FG_PCT', 'OPP_TEAM_FG3_PCT', 'OPP_TEAM_AST', 'OPP_TEAM_REB', 'OPP_TEAM_TOV']
new_column_names = game_general_cols + ['TEAM_ID'] + team_stat_cols + ['OPP_TEAM_ID'] + opp_team_stat_cols

df_gamedays_home = df_games[game_general_cols + game_home_cols + game_away_cols].copy()
# df_gamedays_home = df_games[['GAME_DATE_EST', 'GAME_ID', 'HOME_TEAM_ID', 'PTS_home', 'VISITOR_TEAM_ID', 'PTS_away']].copy()
df_gamedays_home.columns = new_column_names
df_gamedays_home['HOME'] = 1

print(df_gamedays_home.shape)

df_gamedays_away = df_games[game_general_cols + game_away_cols + game_home_cols].copy()
# df_gamedays_away = df_games[['GAME_DATE_EST', 'GAME_ID', 'VISITOR_TEAM_ID', 'PTS_away', 'HOME_TEAM_ID', 'PTS_home']].copy()
df_gamedays_away.columns = new_column_names
df_gamedays_away['HOME'] = 0
print(df_gamedays_away.shape)

## Concatenate the datasets and sort by date/team
df_gamedays = pd.concat([df_gamedays_home, df_gamedays_away], axis=0, ignore_index=True, sort=False).sort_values(by=['GAME_DATE_EST', 'TEAM_ID'])

## add Win column
def win_or_loss(row):
    # if team pts or opp team pts is null then null
    if pd.isnull(row['TEAM_PTS']) or pd.isnull(row['OPP_TEAM_PTS']):
        return np.nan    
    elif row['TEAM_PTS'] > row['OPP_TEAM_PTS']:
        return 1
    else:
        return 0
df_gamedays['WIN'] = df_gamedays.apply(win_or_loss, axis=1)

## remove duplicate game_id and team_id
df_gamedays.drop_duplicates(subset=['GAME_ID', 'TEAM_ID'], inplace=True)

print(df_gamedays.shape)
display(df_gamedays)
```

```python
df_gamedays['GAME_DATE_EST'] = pd.to_datetime(df_gamedays['GAME_DATE_EST'])

## 1. Two features: Get days from previous game and days to next game
df_gamedays['last_game_date'] = df_gamedays.groupby('TEAM_ID')['GAME_DATE_EST'].shift(1)
df_gamedays['days_since_last_game'] = df_gamedays['GAME_DATE_EST'] - df_gamedays['last_game_date']
df_gamedays['days_since_last_game'] = df_gamedays['days_since_last_game'].dt.days.fillna(0).astype(int)

df_gamedays['next_game_date'] = df_gamedays.groupby('TEAM_ID')['GAME_DATE_EST'].shift(-1)
df_gamedays['days_until_next_game'] = df_gamedays['next_game_date'] - df_gamedays['GAME_DATE_EST']
df_gamedays['days_until_next_game'] = df_gamedays['days_until_next_game'].dt.days.fillna(0).astype(int)

## 2. Create two binary features: first_back_to_back, second_back_to_back
# first night of back-to-back if days_since_last_game == 1
df_gamedays['first_back_to_back'] = df_gamedays['days_until_next_game'].apply(lambda x: 1 if x == 1 else 0)

# second night of back-to-back if days_until_next_game == 1
df_gamedays['second_back_to_back'] = df_gamedays['days_since_last_game'].apply(lambda x: 1 if x == 1 else 0)

display(df_gamedays.head())
display(df_gamedays.tail())
```

```python
def per_game_stats_coming_into_game(df, team_cols:t.List, opp_team_cols:t.List) -> pd.DataFrame:
    """
    Get per game stats coming into game for each team

    Args:
        df: dataframe
        team_cols: list of team columns
        opp_team_cols: list of opponent team columns

    Returns:
        dataframe with per game stats coming into game for each team
    """
    df_ = df.copy()
    
    for team_col, opp_col in zip(team_cols, opp_team_cols):
        ## get cumulative average per game and opposing per game for each team up to last game for each season
        df_[f'{team_col}_per_game'] = df_.groupby(['TEAM_ID', 'SEASON'])[team_col].transform(lambda x: x.shift(1).expanding().mean())
        df_[f'{team_col}_opp_per_game'] = df_.groupby(['TEAM_ID', 'SEASON'])[opp_col].transform(lambda x: x.shift(1).expanding().mean())
        
        # ## get opp_team_ppg and opp_team_opp_ppg
        # df_[f'{opp_col}_per_game'] = df_.groupby([f'OPP_TEAM_ID', 'SEASON'])[opp_col].transform(lambda x: x.shift(1).expanding().mean())
        # df_[f'{opp_col}_opp_per_game'] = df_.groupby([f'OPP_TEAM_ID', 'SEASON'])[team_col].transform(lambda x: x.shift(1).expanding().mean())

    return df_

df_gamedays2 = per_game_stats_coming_into_game(df=df_gamedays, team_cols=team_stat_cols, opp_team_cols=opp_team_stat_cols)

print(df_gamedays2.shape)
for i in range(int(len(df_gamedays2.columns)/10)+1):
    try:
        print(i, list(df_gamedays2.columns[i*10:(i+1)*10]))
    except:
        print(i, list(df_gamedays2.columns[i*10:]))
        break
```

```python
## Joins
import pandasql as ps
def query_forecast(game_num:int, df_gamedays_fc: pd.DataFrame, df_team_per_game_expanding: pd.DataFrame) -> pd.DataFrame:
    join_cols = ['HOME', 'days_until_next_game', 'PTS_ALLOWED', 'FG_PCT_ALLOWED', 'FG3_PCT_ALLOWED', 'AST_ALLOWED', 'REB_ALLOWED', 'TOV_FORCED', 'TOV_COMMITTED']
    select_str = ', '.join([f'per_game.{col} as next_{game_num}_{col}' for col in join_cols])

    sqlcode = dedent(f'''
        SELECT fc.*,
            {select_str}
        FROM df_gamedays_fc fc
        LEFT JOIN df_team_per_game_expanding per_game
            ON (fc.next_{game_num}_OPP_TEAM_ID = per_game.TEAM_ID)
                AND (fc.GAME_DATE_EST <= per_game.GAME_DATE_EST)
                AND (fc.GAME_DATE_EST > per_game.LAST_GAME_DATE)
        ;
        ''')

    df = ps.sqldf(sqlcode, locals())
    return df


df_gamedays_fc = df_gamedays2.copy()

## merge df_team_per_game_expanding onto df_gamedays2 on GAME_ID and df_team_per_game_expanding.TEAM_ID = df_gamedays2.next_X_OPP_TEAM_ID for X in range(5), and drop duplicate columns
# df_team_per_game_expanding = df_team_per_game_expanding[['GAME_ID', 'TEAM_ID', 'PTS_ALLOWED', 'FG_PCT_ALLOWED', 'FG3_PCT_ALLOWED', 'AST_ALLOWED', 'REB_ALLOWED', 'TOV_ALLOWED']]
for i in range(5):
    df_gamedays_fc = query_forecast(game_num=i, df_gamedays_fc=df_gamedays_fc, df_team_per_game_expanding=df_team_per_game_expanding)

df_gamedays_fc['GAME_DATE_EST'] = pd.to_datetime(df_gamedays_fc['GAME_DATE_EST'])
df_gamedays_fc
```

```python
def get_min(time_str):
    """Get minutes from time (MM:SS)"""
    if pd.isna(time_str): return np.nan
    try:
        m, s = time_str.split(':')
        mins = float(m) + float(s)/60
    except:
        mins = float(time_str)
    return mins

def game_played(min:float):
    """Binary variable if player played game"""
    # if min is null
    if np.isnan(min):
        return 0
    elif min < 1:
        return 0
    return 1

def comment_cat(comment_str):
    try:
        code, reason = comment_str.split('-', maxsplit=1)
        return code.strip(), reason.strip()
    except:
        return comment_str, comment_str
```

```python
## 3. Join games table onto season table
import pandasql as ps

season_table = "player_seasons" # "player_seasons" or "some_players_seasons"

sqlcode = f'''
    select team.GAME_DATE_EST as staging_game_date_est,
        team.TEAM_ID as staging_team_id,
        team.GAME_ID as staging_game_id,
        team.SEASON as staging_season,
        season.PLAYER_ID as staging_player_id,
        season.PLAYER_NAME as staging_player_name,
        player.*
    from df_gamedays2 team
    inner join {season_table} season
        on team.SEASON = season.SEASON
            and team.TEAM_ID = season.TEAM_ID
            and ((team.TEAM_ID = season.first_team_of_season) or (team.GAME_DATE_EST >= season.min_date))
            and ((team.TEAM_ID = season.last_team_of_season) or (team.GAME_DATE_EST < season.max_date))
    left join df_game_detail_fill_staging player
        on season.PLAYER_ID = player.PLAYER_ID
            and team.GAME_ID = player.GAME_ID
            and team.TEAM_ID = player.TEAM_ID
            and team.SEASON = player.SEASON
            and (
                ((player.next_game_played_date is null) and (team.GAME_DATE_EST > player.GAME_DATE_EST))
                or ((player.last_game_played_date is null) and (team.GAME_DATE_EST <= player.GAME_DATE_EST))
                or ((player.next_game_played_date is not null) and (player.last_game_played_date is not null) and
                    (team.GAME_DATE_EST >= player.GAME_DATE_EST) and (team.GAME_DATE_EST < player.next_game_played_date))
            )
    ;
    '''

players_games_filled = ps.sqldf(sqlcode,locals())
display(players_games_filled[(players_games_filled['staging_player_name'] == 'Cade Cunningham')].head(6))
```
