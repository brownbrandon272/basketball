## Initial Feature Engineering
Converting minutes to numeric and quick feature engineering for the comments field

```python
df2 = df_game_detail.copy()
df2.drop(['GAME_ID', 'TEAM_ID', 'TEAM_CITY', 'TEAM_ABBREVIATION', 'PLAYER_ID', 'PLAYER_NAME', 'NICKNAME'], axis=1, inplace=True)
```

```python
def get_min(time_str):
    """Get minutes from time (MM:SS)"""
    try:
        m, s = time_str.split(':')
        mins = float(m) + float(s)/60
    except:
        mins = float(time_str)
    return mins

df2['MIN'] = df2['MIN'].apply(get_min)
print(df2['MIN'].dtype)
```

```python
pd.set_option("display.max_rows", None)
print(df2['COMMENT'].value_counts())
pd.set_option("display.max_rows", 30)
```

```python
def comment_cat(comment_str):
    try:
        code, reason = comment_str.split('-', maxsplit=1)
        return code.strip(), reason.strip()
    except:
        return comment_str, comment_str

df2['comment_code'], df2['comment_reason'] = zip(*df2['COMMENT'].apply(comment_cat))
```

```python
pd.set_option("display.max_rows", None)
print(df2['comment_code'].value_counts(), "\n")
print(df2['comment_reason'].value_counts())
pd.set_option("display.max_rows", 30)
```

```python
## Change to per minute stats
cols_to_make_per_min = \
    ['FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA', 'OREB', 'DREB',
    'REB', 'AST', 'STL', 'BLK', 'TO', 'PTS', 'PLUS_MINUS', 'PF',
    ]

per_min_cols = [x + '_pm' for x in cols_to_make_per_min]
```

```python
arr1 = np.array(df2[cols_to_make_per_min])
arr2 = np.array(df2['MIN'])

df_per_min = pd.DataFrame((arr1.T / arr2).T, columns=per_min_cols)
df3 = pd.concat([df2, df_per_min], axis=1)
df3
```

```python
var_dict = {
    'num_vars':
        ['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT',
        'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TO', 'PTS', 'PLUS_MINUS', 'MIN', 'PF',
        ],
    'cat_vars':
        ['START_POSITION', 'comment_code', 'comment_reason']
}
```

```python
for c in df2.columns:
    if c in var_dict['num_vars']:
        df2[c] = pd.to_numeric(df2[c], errors='raise')
        print(f"{c} data type: {df2[c].dtype}")
    elif c in var_dict['cat_vars']:
        print(f"{c} data type: {df2[c].dtype}")
    else:
        print(f'column {c} not in var_dict')
```

```python
# 0. Transform games table from HOME_TEAM, AWAY_TEAM to TEAM, HOME/AWAY 

df_gamedays_home = df_games[['GAME_DATE_EST', 'GAME_ID', 'HOME_TEAM_ID', 'PTS_home']].copy()
df_gamedays_home.columns = ['GAME_DATE_EST', 'GAME_ID', 'TEAM_ID', 'PTS']
df_gamedays_home['HOME'] = 1
print(df_gamedays_home.shape)

df_gamedays_away = df_games[['GAME_DATE_EST', 'GAME_ID', 'VISITOR_TEAM_ID', 'PTS_away']].copy()
df_gamedays_away.columns = ['GAME_DATE_EST', 'GAME_ID', 'TEAM_ID', 'PTS']
df_gamedays_away['HOME'] = 0
print(df_gamedays_away.shape)

# 1. Order games by team and date
df_gamedays = pd.concat([df_gamedays_home, df_gamedays_away], axis=0, ignore_index=True, sort=False).sort_values(by=['TEAM_ID', 'GAME_DATE_EST'])
print(df_gamedays.shape)
display(df_gamedays)
```

```python
# 2A. Find consecutive GAME_DATE_EST that are more than a month apart
df_gamedays['GAME_DATE_EST'] = pd.to_datetime(df_gamedays['GAME_DATE_EST'])

df_gamedays['last_game_date'] = df_gamedays.groupby('TEAM_ID')['GAME_DATE_EST'].shift(1)
df_gamedays['days_since_last_game'] = df_gamedays['GAME_DATE_EST'] - df_gamedays['last_game_date']
df_gamedays['days_since_last_game'] = df_gamedays['days_since_last_game'].dt.days.fillna(0).astype(int)

df_gamedays['next_game_date'] = df_gamedays.groupby('TEAM_ID')['GAME_DATE_EST'].shift(-1)
df_gamedays['days_until_next_game'] = df_gamedays['next_game_date'] - df_gamedays['GAME_DATE_EST']
df_gamedays['days_until_next_game'] = df_gamedays['days_until_next_game'].dt.days.fillna(0).astype(int)

# first night of back-to-back if days_since_last_game == 1
df_gamedays['first_back_to_back'] = df_gamedays['days_until_next_game'].apply(lambda x: 1 if x == 1 else 0)

# second night of back-to-back if days_until_next_game == 1
df_gamedays['second_back_to_back'] = df_gamedays['days_since_last_game'].apply(lambda x: 1 if x == 1 else 0)

display(df_gamedays.head())
display(df_gamedays.tail())
```

```python
import pandasql as ps

sqlcode = '''
select gd.*, s.season, s.season_short, s.min_games, s.max_games, s.number_of_teams, s.preseason_end, s.postseason_start, s.postseason_end
from df_gamedays gd
inner join df_seasons s on gd.GAME_DATE_EST>= s.season_start and gd.GAME_DATE_EST <= s.season_end
'''

df_gamedays2 = ps.sqldf(sqlcode,locals())

## make datetime format yyyy-mm-dd: GAME_DATE_EST, date_lag, preseason_end, postseason_start, postseason_end
df_gamedays2['GAME_DATE_EST'] = pd.to_datetime(df_gamedays2['GAME_DATE_EST'])
df_gamedays2['date_lag'] = pd.to_datetime(df_gamedays2['date_lag'])
df_gamedays2['preseason_end'] = pd.to_datetime(df_gamedays2['preseason_end'])
df_gamedays2['postseason_start'] = pd.to_datetime(df_gamedays2['postseason_start'])
df_gamedays2['postseason_end'] = pd.to_datetime(df_gamedays2['postseason_end'])

display(df_gamedays2.iloc[[0, 1, 2, 53299, 53300, 53301]])
```

```python
# regular season if between preseason_end and postseason_start
df_gamedays2['regular_season'] = df_gamedays2.apply(lambda x: 1 if x['GAME_DATE_EST'] >= x['preseason_end'] and x['GAME_DATE_EST'] < x['postseason_start'] else 0, axis=1)

# preseason if before preseason_end
df_gamedays2['preseason'] = df_gamedays2.apply(lambda x: 1 if x['GAME_DATE_EST'] < x['preseason_end'] else 0, axis=1)

# postseason if between postseason_start and postseason_end
df_gamedays2['postseason'] = df_gamedays2.apply(lambda x: 1 if x['GAME_DATE_EST'] >= x['postseason_start'] and x['GAME_DATE_EST'] <= x['postseason_end'] else 0, axis=1)

# get game number of season for team
df_gamedays2['game_number_of_season'] = df_gamedays2.groupby(['TEAM_ID', 'season'])['regular_season'].cumsum()

# get total number of regular_season games played by team in season, no less than of min_games
df_gamedays2['total_regular_season_games'] = df_gamedays2.groupby(['TEAM_ID', 'season'])['regular_season'].transform('sum')
df_gamedays2['total_regular_season_games'] = df_gamedays2['total_regular_season_games'].apply(lambda x: x if x >= df_gamedays2['min_games'].iloc[0] else df_gamedays2['min_games'].iloc[0])

# percentage of regaular season games played - game number divided by max of total regular season games and min_games
df_gamedays2['percentage_regular_season_games'] = df_gamedays2['game_number_of_season'] / df_gamedays2['total_regular_season_games']

display(df_gamedays2.iloc[[0, 1, 2, 53299, 53300, 53301]])
```

```python
# 3. Get rolling averages
def get_lags(df, col, n_lags, prefix):
    for i in range(1, n_lags + 1):
        df[f'{prefix}{i}'] = df.groupby(['TEAM_ID', 'season'])[col].shift(i)
    return df

def get_average_windows(df, col, n_lags, prefix):
    for i in range(1, n_lags + 1):
        df[f'{prefix}{i}'] = df.groupby(['TEAM_ID', 'season'])[col].shift(i).rolling(i).mean()
    return df
```
