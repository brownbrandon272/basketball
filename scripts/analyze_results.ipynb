```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games

# timeseries plot
def plot_timeseries(player_name:str, season:str, yhat_df:pd.DataFrame, y_test:pd.DataFrame, idx_test:pd.DataFrame):
    # get the index of the player id and date
    idx = idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)].index[0]
    # get the next 5 games
    yhat_df_slice = yhat_df.iloc[idx:idx+5]
    y_test_slice = y_test.iloc[idx:idx+5]
    # plot
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.plot(yhat_df_slice.index, yhat_df_slice['next_0_MIN'], label='Predicted')
    ax.plot(y_test_slice.index, y_test_slice['next_0_MIN'], label='Actual')
    ax.set_title('Minutes Played')
    ax.set_xlabel('Game')
    ax.set_ylabel('Minutes Played')
    ax.legend()
    plt.show()

    # fig, ax = plt.subplots(figsize=(15, 5))
    # ax.plot(yhat_df_slice.index, yhat_df_slice['next_0_PTS'], label='Predicted')
    # ax.plot(y_test_slice.index, y_test_slice['next_0_PTS'], label='Actual')
    # ax.set_title('Points')
    # ax.set_xlabel('Game')
    # ax.set_ylabel('Points')
    # ax.legend()
    # plt.show()


player_name = 'Nikola Jokic'
season = '2021-2022'
plot_timeseries(player_name=player_name, season=season, yhat_df=yhat_df, y_test=y_test, idx_test=idx_test)
# # idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)].index

# (idx_test['PLAYER_NAME'] == player_name).sum()
# (idx_test['season'] == season).sum()
```
