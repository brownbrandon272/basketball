```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games

# timeseries plot
def plot_timeseries(player_name:str, season:str, yhat_df:pd.DataFrame, y_test:pd.DataFrame, idx_test:pd.DataFrame):
    # get the index of the player id and date
    idx = idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)].index[0]
    # get the next 5 games
    yhat_df_slice = yhat_df.iloc[idx:idx+5]
    y_test_slice = y_test.iloc[idx:idx+5]
    # plot
    fig, ax = plt.subplots(figsize=(15, 5))
    ax.plot(yhat_df_slice.index, yhat_df_slice['next_0_MIN'], label='Predicted')
    ax.plot(y_test_slice.index, y_test_slice['next_0_MIN'], label='Actual')
    ax.set_title('Minutes Played')
    ax.set_xlabel('Game')
    ax.set_ylabel('Minutes Played')
    ax.legend()
    plt.show()

    # fig, ax = plt.subplots(figsize=(15, 5))
    # ax.plot(yhat_df_slice.index, yhat_df_slice['next_0_PTS'], label='Predicted')
    # ax.plot(y_test_slice.index, y_test_slice['next_0_PTS'], label='Actual')
    # ax.set_title('Points')
    # ax.set_xlabel('Game')
    # ax.set_ylabel('Points')
    # ax.legend()
    # plt.show()


player_name = 'Nikola Jokic'
season = '2021-2022'
plot_timeseries(player_name=player_name, season=season, yhat_df=yhat_df, y_test=y_test, idx_test=idx_test)
# # idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)].index

# (idx_test['PLAYER_NAME'] == player_name).sum()
# (idx_test['season'] == season).sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# create temp_df
temp_df = pd.DataFrame(
    {'next_0_MIN': {0: 19.0, 3: 28.0},
     'next_1_MIN': {0: 17.0, 3: 28.0},
     'next_2_MIN': {0: 17.0, 3: 27.0},
     'next_0_DATE': {0: '2022-03-14', 3: '2022-03-22'},
     'next_1_DATE': {0: '2022-03-16', 3: '2022-03-23'},
     'next_2_DATE': {0: '2022-03-19', 3: '2022-03-26'},
    })

# create temp_df2
temp_df2 = pd.DataFrame(
    {'next_0_MIN': {0: 21.2, 3: 24.0},
     'next_1_MIN': {0: 18.1, 3: 27.9},
     'next_2_MIN': {0: 16.8, 3: 26.3},
    })

# create a one dimensional array of values
values = np.array([])
values2 = np.array([])
idxes = np.array([])
dates = np.array([])
min_cols = [col for col in temp_df.columns if 'MIN' in col]
date_cols = [col for col in temp_df.columns if 'DATE' in col]
for idx in temp_df2.index:
    for col in min_cols:
        values = np.append(values, temp_df.loc[idx, col])
        values2 = np.append(values2, temp_df2.loc[idx, col])
        idxes = np.append(idxes, idx)
    for col in date_cols:
        dates = np.append(dates, temp_df.loc[idx, col])
dates = pd.to_datetime(dates)

## create dataframe out of values, values2, idxes, dates
df = pd.DataFrame({'values': values, 'values2': values2, 'idxes': idxes, 'dates': dates})
display(df.to_dict())

# Step 4
fig, ax = plt.subplots()
ax.scatter(dates, values, marker='o', label='temp_df')
ax.plot(dates, values2, marker='o', label='temp_df2', color='orange')

# Step 5
ax.set_xlabel('Date')
ax.set_ylabel('Minutes')
days = mdates.DayLocator()  # Set the tick locator to show days
ax.xaxis.set_major_locator(days)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Set the x-axis tick labels to YYYY-MM-DD format
fig.autofmt_xdate(rotation=45)  # Rotate x-axis tick labels for better visibility

# show plot
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create the DataFrame
df = pd.DataFrame({'values': {0: 19.0, 1: 17.0, 2: 17.0, 3: 28.0, 4: 28.0, 5: 27.0},
                   'values2': {0: 21.2, 1: 18.1, 2: 16.8, 3: 24.0, 4: 27.9, 5: 26.3},
                   'idxes': {0: 0.0, 1: 0.0, 2: 0.0, 3: 3.0, 4: 3.0, 5: 3.0},
                   'dates': {0: pd.Timestamp('2022-03-14 00:00:00'),
                             1: pd.Timestamp('2022-03-16 00:00:00'),
                             2: pd.Timestamp('2022-03-19 00:00:00'),
                             3: pd.Timestamp('2022-03-22 00:00:00'),
                             4: pd.Timestamp('2022-03-23 00:00:00'),
                             5: pd.Timestamp('2022-03-26 00:00:00')}
                  })

# create the figure and axis objects
fig, ax = plt.subplots()

# create scatter plot of 'values' column with dates on x-axis, and set all points to the same color
ax.scatter(df['dates'], df['values'], c='C0', label='Actual')

# loop through unique 'idxes' values in 'values2', and connect the corresponding points with a line of the same color
for idx in df['idxes'].unique():
    idx_rows = df[df['idxes'] == idx]
    ax.plot(idx_rows['dates'], idx_rows['values2'], c='C3', label='Forecast' if idx == df['idxes'].max() else None) # c='C{}'.format(int(idx))

# Step 5
ax.set_xlabel('Date')
ax.set_ylabel('Minutes')
days = mdates.DayLocator()  # Set the tick locator to show days
ax.xaxis.set_major_locator(days)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Set the x-axis tick labels to YYYY-MM-DD format
fig.autofmt_xdate(rotation=45)  # Rotate x-axis tick labels for better visibility

ax.legend()

# show the plot
plt.show()
```
