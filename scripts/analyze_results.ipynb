```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games

def get_all_lags(df:pd.DataFrame, cols:t.List, group_cols:t.List, n_lags:int, order_cols:t.List=None) -> t.Tuple[pd.DataFrame, t.List[str]]:
    df_ = df.copy()
    if order_cols is not None: df_ = df_.sort_values(by=order_cols)
    created_cols = []

    if n_lags > 0:
        start, stop, step = 1, n_lags + 1, 1
        prefix = 'lag_'
    else:
        start, stop, step = 0, n_lags, -1
        prefix = 'next_'
    
    # for each column in cols, get n_lags number of lags
    for col in cols:
        # for each lag, create a new column with the lagged value
        for i in range(start, stop, step):
            created_col = f'{prefix}{abs(i)}_{col}'
            created_cols.append(created_col)
            df_[created_col] = df_.groupby(group_cols)[col].shift(i)

    return df_, created_cols
```

```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games
import pandas as pd
import matplotlib.dates as mdates

def plot_timeseries(player_name, season, first_game_i=10, last_game_i=30, every_n_games=3, col_ends = ['PTS', 'PTS_PRED', 'GAME_DATE_EST']):
    df_staging = idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)] # .index

    # join y_test onto df_staging
    df_staging = df_staging.merge(y_test, how='inner', left_index=True, right_index=True)

    # join yhat_df onto df_staging
    df_staging = df_staging.merge(yhat_df, how='inner', left_index=True, right_index=True, suffixes=('', '_PRED'))

    # select columns either in idx_test or containing '_PTS_'
    df = df_staging[[col for col in df_staging.columns if col in idx_test.columns or '_PTS' in col]].copy()
    df['GAME_DATE_EST'] = pd.to_datetime(df['GAME_DATE_EST']).dt.date

    # get next 5 GAME_DATE_EST using get_all_lags()
    df, _ = get_all_lags(df, cols=['GAME_DATE_EST'], group_cols=['PLAYER_NAME', 'season'], n_lags=-5, order_cols=['GAME_DATE_EST'])

    all_plot_cols = []
    all_plot_col_tuples = []

    for col_end in col_ends:
        plot_cols = [col for col in df.columns if col.startswith('next') and col.endswith(col_end)]
        all_plot_cols.extend(plot_cols)
        plot_col_tuples = [(int(col.split('_')[1]), col_end) for col in plot_cols]
        all_plot_col_tuples.extend(plot_col_tuples)

    df = df[all_plot_cols]
    multi_cols = pd.MultiIndex.from_tuples(all_plot_col_tuples, names=['Game', 'Metric'])

    df.columns = multi_cols
    df2 = df.stack(level='Game')
    df2.dropna(inplace=True)

    fig, ax = plt.subplots()

    plot_dates = []
    first = True
    for i, (idx, new_df) in enumerate(df2.groupby(level=0)):
        if i < first_game_i: continue
        if i > last_game_i: break
        if i % every_n_games != 0: continue
        plot_dates.extend(new_df['GAME_DATE_EST'].values)
        ax.scatter(x=new_df['GAME_DATE_EST'].values, y=new_df['PTS'].values, c='C0', label='Actual' if first else None)
        ax.plot(new_df['GAME_DATE_EST'].values, new_df['PTS_PRED'].values, c='C3', marker='o', label='Forecast' if first else None) # c='C{}'.format(int(idx))
        first = False


    # Step 5
    ax.set_xlabel('Date')
    ax.set_ylabel('Points')
    days = mdates.DayLocator()  # Set the tick locator to show days
    ax.xaxis.set_major_locator(days)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Set the x-axis tick labels to YYYY-MM-DD format
    fig.autofmt_xdate(rotation=45)  # Rotate x-axis tick labels for better visibility

    ax.legend()

    ## create custom labels for x-axis: if date exists then date, else ''
    max_date = max(plot_dates) + pd.Timedelta(days=1)
    min_date = min(plot_dates) - pd.Timedelta(days=1)
    labels = pd.date_range(start=min_date, end=max_date, freq='D')

    # remove x ticks that are less than min_date or greater than max_date
    ax.set_xlim(left=min_date, right=max_date)

    # # add some white space to the left and right of the plot
    # ax.set_xlim(left=min_date - pd.Timedelta(days=1), right=max_date + pd.Timedelta(days=1))

    custom_labels = [dt.strftime('%Y-%m-%d') if dt.date() in plot_dates else '' for dt in labels]

    # set the custom labels
    ax.set_xticklabels(custom_labels)

    ## adjust figure size
    fig.set_size_inches(14, 4)

    ## set title
    ax.set_title(f'{player_name} - {season} - Actual vs. Predicted Points')

    # show the plot
    plt.show()
```

```python
PLAYER_NAME = 'Anthony Davis'
SEASON = '2021-2022'
FIRST_GAME_I = 40
LAST_GAME_I = 82
EVERY_N_GAMES = 3
COL_ENDS = ['PTS', 'PTS_PRED', 'GAME_DATE_EST']

plot_timeseries(player_name=PLAYER_NAME,
                season=SEASON,
                first_game_i=FIRST_GAME_I,
                last_game_i=LAST_GAME_I,
                every_n_games=EVERY_N_GAMES,
                col_ends=COL_ENDS)
```
