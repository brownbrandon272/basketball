```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games

def get_all_lags(df:pd.DataFrame, cols:t.List, group_cols:t.List, n_lags:int, order_cols:t.List=None) -> t.Tuple[pd.DataFrame, t.List[str]]:
    df_ = df.copy()
    if order_cols is not None: df_ = df_.sort_values(by=order_cols)
    created_cols = []

    if n_lags > 0:
        start, stop, step = 1, n_lags + 1, 1
        prefix = 'lag_'
    else:
        start, stop, step = 0, n_lags, -1
        prefix = 'next_'
    
    # for each column in cols, get n_lags number of lags
    for col in cols:
        # for each lag, create a new column with the lagged value
        for i in range(start, stop, step):
            created_col = f'{prefix}{abs(i)}_{col}'
            created_cols.append(created_col)
            df_[created_col] = df_.groupby(group_cols)[col].shift(i)

    return df_, created_cols


player_name = 'Nikola Jokic'
season = '2021-2022'
jokic2022 = idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)] # .index

# join y_test onto jokic2022
jokic2022 = jokic2022.merge(y_test, how='inner', left_index=True, right_index=True)

# join yhat_df onto jokic2022
jokic2022 = jokic2022.merge(yhat_df, how='inner', left_index=True, right_index=True, suffixes=('', '_PRED'))

# select columns either in idx_test or containing '_PTS_'
plot_jokic = jokic2022[[col for col in jokic2022.columns if col in idx_test.columns or '_PTS' in col]].copy()
plot_jokic['GAME_DATE_EST'] = pd.to_datetime(plot_jokic['GAME_DATE_EST']).dt.date
plot_jokic

# get next 5 GAME_DATE_EST using get_all_lags()
plot_jokic, _ = get_all_lags(plot_jokic, cols=['GAME_DATE_EST'], group_cols=['PLAYER_NAME', 'season'], n_lags=-5, order_cols=['GAME_DATE_EST'])

# display(plot_jokic.head(2))
# print(plot_jokic.dtypes)

# import matplotlib.dates as mdates

date_cols = [col for col in plot_jokic.columns if ('GAME_DATE_EST' in col) and ('next' in col)]
# convert date_cols to datetime
plot_jokic[date_cols] = plot_jokic[date_cols].apply(lambda x: pd.to_datetime(x).dt.date)

actual_pts_cols = [col for col in plot_jokic.columns if ('PTS' in col) and ('PRED' not in col)]
pred_pts_cols = [col for col in plot_jokic.columns if ('PTS' in col) and ('PRED' in col)]

fig, ax = plt.subplots()

for i, (idx, x) in enumerate(plot_jokic.iterrows()):
    if i%2 != 0: continue
    date_x = x[pd.notna(x).values & x.index.isin(date_cols)].values
    actual_y = x[pd.notna(x).values & x.index.isin(actual_pts_cols)].values
    pred_y = x[pd.notna(x).values & x.index.isin(pred_pts_cols)].values
    print(date_x, actual_y, pred_y)
    ax.scatter(x=date_x, y=actual_y, c='C0', label='Actual' if idx == plot_jokic.index.min() else None)
    ax.plot(date_x, pred_y, c='C3', label='Forecast' if idx == plot_jokic.index.min() else None) # c='C{}'.format(int(idx))
    
# Step 5
ax.set_xlabel('Date')
ax.set_ylabel('Points')
days = mdates.DayLocator()  # Set the tick locator to show days
ax.xaxis.set_major_locator(days)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Set the x-axis tick labels to YYYY-MM-DD format
fig.autofmt_xdate(rotation=45)  # Rotate x-axis tick labels for better visibility

ax.legend()

## create custom labels for x-axis: if date exists then date, else ''
max_date = plot_jokic['GAME_DATE_EST'].max()
min_date = plot_jokic['GAME_DATE_EST'].min()
labels = pd.date_range(start=min_date, end=max_date, freq='D')

custom_labels = [dt.strftime('%Y-%m-%d') if dt in plot_jokic['GAME_DATE_EST'].values else '' for dt in labels]

# set the custom labels
ax.set_xticklabels(custom_labels)

## adjust figure size
fig.set_size_inches(14, 4)



# show the plot
plt.show()
```
