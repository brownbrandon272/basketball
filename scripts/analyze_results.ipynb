```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games

def get_all_lags(df:pd.DataFrame, cols:t.List, group_cols:t.List, n_lags:int, order_cols:t.List=None) -> t.Tuple[pd.DataFrame, t.List[str]]:
    df_ = df.copy()
    if order_cols is not None: df_ = df_.sort_values(by=order_cols)
    created_cols = []

    if n_lags > 0:
        start, stop, step = 1, n_lags + 1, 1
        prefix = 'lag_'
    else:
        start, stop, step = 0, n_lags, -1
        prefix = 'next_'
    
    # for each column in cols, get n_lags number of lags
    for col in cols:
        # for each lag, create a new column with the lagged value
        for i in range(start, stop, step):
            created_col = f'{prefix}{abs(i)}_{col}'
            created_cols.append(created_col)
            df_[created_col] = df_.groupby(group_cols)[col].shift(i)

    return df_, created_cols
```

```python
## create a function that takes a player id and date and creates a timeseries plot comparing the actual and predicted values for the next 5 games
import pandas as pd
import matplotlib.dates as mdates

def plot_timeseries(player_name, season, first_game_i=10, last_game_i=30, every_n_games=3, col_ends=['PTS']):
    date_field = 'GAME_DATE_EST'
    df_staging = idx_test[(idx_test['PLAYER_NAME'] == player_name) & (idx_test['season'] == season)]

    df_staging = df_staging.merge(y_test, how='inner', left_index=True, right_index=True)
    df_staging = df_staging.merge(yhat_df, how='inner', left_index=True, right_index=True, suffixes=('', '_PRED'))

    df = df_staging[[col for col in df_staging.columns if col in idx_test.columns or any('_' + x in col for x in col_ends)]].copy()
    df['GAME_DATE_EST'] = pd.to_datetime(df['GAME_DATE_EST']).dt.date

    df, _ = get_all_lags(df, cols=['GAME_DATE_EST'], group_cols=['PLAYER_NAME', 'season'], n_lags=-5, order_cols=[date_field])

    all_plot_cols = []
    all_plot_col_tuples = []
    all_col_ends = set([date_field] + col_ends + [x + '_PRED' for x in col_ends])

    for col_end in all_col_ends:
        plot_cols = [col for col in df.columns if col.startswith('next') and col.endswith(col_end)]
        all_plot_cols.extend(plot_cols)
        plot_col_tuples = [(int(col.split('_')[1]), col_end) for col in plot_cols]
        all_plot_col_tuples.extend(plot_col_tuples)

    df = df[all_plot_cols]
    multi_cols = pd.MultiIndex.from_tuples(all_plot_col_tuples, names=['Game', 'Metric'])

    df.columns = multi_cols
    df2 = df.stack(level='Game')
    df2.dropna(inplace=True)

    fig, axs = plt.subplots(len(col_ends), 1, sharex=True, figsize=(14, 4*len(col_ends)))

    plot_dates = []
    first = True

    for i, (idx, new_df) in enumerate(df2.groupby(level=0)):
        if i < first_game_i: continue
        if i > last_game_i: break
        if i % every_n_games != 0: continue
        plot_dates.extend(new_df[date_field].values)

        for j, col_end in enumerate(col_ends):
            ax = axs[j]
            ax.scatter(x=new_df[date_field].values, y=new_df[col_end].values, c='C0', label='Actual' if first else None)
            ax.plot(new_df[date_field].values, new_df[col_end + '_PRED'].values, c='C3', marker='o', label='Forecast' if first else None)
            first = False

    for ax, col_end in zip(axs, col_ends):
        ax.set_ylabel(col_end)
        # ax.legend()

    axs[-1].set_xlabel('Date')
    days = mdates.DayLocator()
    axs[-1].xaxis.set_major_locator(days)
    axs[-1].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    fig.autofmt_xdate(rotation=45)

    max_date = max(plot_dates) + pd.Timedelta(days=1)
    min_date = min(plot_dates) - pd.Timedelta(days=1)
    labels = pd.date_range(start=min_date, end=max_date, freq='D')

    axs[-1].set_xlim(left=min_date, right=max_date)

    custom_labels = [dt.strftime('%Y-%m-%d') if dt.date() in plot_dates else '' for dt in labels]
    
    axs[-1].set_xticks(labels)
    axs[-1].set_xticklabels(custom_labels)

    fig.suptitle(f'{player_name} - {season} - Actual vs. Predicted Values')
    
    # Move the legend to the top of the first subplot
    axs[0].legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=2)

    plt.tight_layout()
    plt.show()
```

```python
PLAYER_NAME = 'Cade Cunningham'
SEASON = '2021-2022'
FIRST_GAME_I = 0
LAST_GAME_I = 40
EVERY_N_GAMES = 3
COL_ENDS = ['PTS', 'REB']

plot_timeseries(player_name=PLAYER_NAME,
                season=SEASON,
                first_game_i=FIRST_GAME_I,
                last_game_i=LAST_GAME_I,
                every_n_games=EVERY_N_GAMES,
                col_ends=COL_ENDS)
```
